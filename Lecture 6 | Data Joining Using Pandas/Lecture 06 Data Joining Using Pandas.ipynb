```python
import pandas as pd
```

### Create two random dataframes

```python
df_A = pd.DataFrame()
df_A["Name"] = ["A","B","C","D"]  
df_A["Address"] = ["Budapest","Dhaka","Seoul", "Waterloo"]
df_A["University"] = ["BME","DU","SNU","UoW"]

print(type(df_A))
print(df_A.shape)
display(df_A)
```

```python
df_B = {'Name': ['A', 'Y', 'X', 'B'], 
        'Department': ["E","F","G","H"],
        'Address':["I","J","K","L"]}

df_B = pd.DataFrame(df_B)
display(df_B)
```

### Inner Join 

```python
df_Inner_E1 = df_A.merge(df_B, how = "inner", on = ["Name"], suffixes=["_A","_B"])

display(df_Inner_E1)
```

```python
df_D = pd.DataFrame()
df_D["Name"] = ["A","B","C","D"]  
df_D["Address"] = ["Budapest","Dhaka","Seoul", "Waterloo"]
df_D["University"] = ["BME","DU","SNU","UoW"]

df_C = {'Name': ['A', 'Y', 'X', 'B'], 
        'Department': ["E","F","G","H"],
        'Address':["NYC","J","K","Dhaka"]}

df_C = pd.DataFrame(df_C)

df_Inner_E2 = df_D.merge(df_C, how = "inner", on = ["Name"])

df_Inner_E3 = df_D.merge(df_C, how = "inner", on = ["Address"])

df_Inner_E4 = df_D.merge(df_C, how = "inner", on = ["Name","Address"])

display(df_Inner_E2)
display(df_Inner_E3)
display(df_Inner_E4)
```

### What if we change?

```python
df_Inner_E5 = df_C.merge(df_D, how = "inner", on = ["Name"])
display(df_Inner_E5)
```

### Left Join 

```python
display(df_A)
display(df_C)
```

```python
df_left_e1 = df_A.merge(df_C, how="left")
df_left_e2 = df_A.merge(df_C, how="left", on=["Name", "Address"])
df_left_e3 = df_A.merge(df_C, how="left", on=["Name"])
df_left_e4 = df_A.merge(df_C, how="left", on=["Address"])

display(df_left_e1)
display(df_left_e2)
display(df_left_e3)
display(df_left_e4)
```

### Outer Join 

```python
display(df_A)
display(df_C)
```

```python
df_outer_e1 = df_A.merge(df_C, how="outer")
df_outer_e2 = df_A.merge(df_C, how="outer", on=["Name", "Address"])
df_outer_e3 = df_A.merge(df_C, how="outer", on=["Name"])
df_outer_e4 = df_A.merge(df_C, how="outer", on=["Address"])

display(df_outer_e1)
display(df_outer_e2)
display(df_outer_e3)
display(df_outer_e4)
```

### Right Join

```python
display(df_A)
display(df_C)
```

```python
df_right_e1 = df_A.merge(df_C, how="right")
df_right_e2 = df_A.merge(df_C, how="right", on=["Name", "Address"])
df_right_e3 = df_A.merge(df_C, how="right", on=["Name"])
df_right_e4 = df_A.merge(df_C, how="right", on=["Address"])

display(df_right_e1)
display(df_right_e2)
display(df_right_e3)
display(df_right_e4)
```
