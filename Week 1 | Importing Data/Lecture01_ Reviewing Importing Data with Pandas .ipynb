### Reading the csv file in a data frame and demonstrate the data frame structures

```python
#reading a csv file
df= pd.read_csv("batsman_most_runs_ODI.csv")

print(type(df)) #type of the data frame
print("____________________________________________________________")
display(df.head()) #first 5 rows
print("____________________________________________________________")
display(df.tail(15))#last 15 rows
print("____________________________________________________________")
print(df.shape)#to find the number of rows and columns
print("number of rows= ",df.shape[0],", number of columns=",df.shape[1])
```

```python
#checking for missing values and data types of each column
print(df.info())
```
