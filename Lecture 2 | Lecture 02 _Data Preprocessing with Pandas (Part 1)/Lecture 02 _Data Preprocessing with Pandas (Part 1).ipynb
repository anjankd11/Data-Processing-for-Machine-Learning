### Reading Excel file

```python
#df = pd.read_excel("ODI_cricket.xlsx", sheet_name="bowler")
df = pd.read_excel("ODI_cricket.xlsx", sheet_name="batsman")
#engine="openpyxl"
#pip install openpyxl
display(df.head())
```

### How to rename columns

```python
print(df.columns)
```

```python
#use dictionary
df=df.rename(columns={'Mat':'Match', 
                        'Inns':'Innings',
                        'NO': 'NotOut',
                        'HS': 'Highest_score',
                        'Ave': 'Average',
                        'BF': 'Balls_Faced',
                        'SR': 'Strike_Rate',
                        100: 'Centuries',
                        50: 'Half_centuries',
                        0: 'Ducks',
                        "4s": "Fours",
                        "6s": "Sixes"
    
})
display(df.head())
```

### How to split a column and create two new columns?

```python
df['Player']
```

```python
df_player = df['Player'].str.split("(")
print(df_player)
```

```python
df_player = df['Player'].str.split("(", expand=True)

display(df_player)
```

```python
df[["Player_Name", "Country"]] = df['Player'].str.split("(", expand=True)

display(df.head())
```

### How to remove a column

```python
# method 1
df = df.drop('Player', axis=1)

# method 2
#df.drop('Player', axis=1, inplace=True)

display(df.head())
```

### How to replace/remove a value from a pandas column?

```python
df['Country'] = df['Country'].str.replace(")", "")

display(df.head())
```

```python
print(df.columns)

new_col_sequence = ['Player_Name', 'Country', 'Span', 'Match', 'Innings', 'NotOut', 'Runs', 'Highest_score',
       'Average', 'Balls_Faced', 'Strike_Rate', 'Centuries', 'Half_centuries', 'Ducks', 'Fours', 'Sixes']
```

```python
df=df[new_col_sequence]
display(df.head())
```

Creating local file of the modified csv

```python
df.to_csv("processed_ODI_batsman.csv", index=False)
```
